```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('datasets/RawData.csv')
```

```python
# Find the duplicate occurrences of ClientID and CycleNumber
duplicate_occurrences = data[data.duplicated(['ClientID', 'CycleNumber'], keep=False)]

# Group the duplicate occurrences by ClientID and CycleNumber and count the occurrences
duplicate_counts = duplicate_occurrences.groupby(['ClientID', 'CycleNumber']).size().reset_index(name='OccurrenceCount')

# Filter the duplicate counts to keep the occurrences with count greater than 1
duplicate_counts = duplicate_counts[duplicate_counts['OccurrenceCount'] > 1]

# Print the duplicate occurrences
if duplicate_counts.empty:
    print("No occurrences with multiple rows found.")
else:
    print("Occurrences with multiple rows:")
    print(duplicate_counts)
```

```python
to_remove_list = [x for x in duplicate_counts["ClientID"].unique() if x !='nfp8107']
to_remove_list
```

```python
len(data)
```

```python
data = data[~data["ClientID"].isin(to_remove_list)]
data.reset_index(drop=True, inplace=True)
```

```python
data
```

```python
filtered_df = data.groupby('ClientID').filter(lambda x: len(x) > 4)
filtered_df.reset_index(drop=True, inplace=True)
```
